```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scale_cols = ['JavaScript', 'Shell', 'HTML', 'Ruby', 'CSS', 'Python', 'C', 'Java', 'C++', 'PHP']
df_scaled = scaler.fit_transform(fdf[scale_cols])

df_scaled = pd.DataFrame(df_scaled)
df_scaled.columns = scale_cols

print(df_scaled)
```

```python
Test_SIZE = 90
train = df_scaled[:Test_SIZE]
test = df_scaled[Test_SIZE:]
```

```python
def make_dataset(data, label, window_size=20):
    feature_list = []
    label_list = []
    for i in range(len(data) - window_size):
        feature_list.append(np.array(data.iloc[i:i+window_size]))
        label_list.append(np.array(label.iloc[i+window_size]))
    return np.array(feature_list), np.array(label_list)
```

```python
feature_cols = ['JavaScript','Shell', 'HTML', 'Ruby',  'Python', 'C', 'Java', 'C++', 'PHP']
label_cols = ['CSS']

train_feature = train[feature_cols]
train_label = train[label_cols]

train_feature, train_label = make_dataset(train_feature, train_label, 20)
from sklearn.model_selection import train_test_split
x_train,  x_valid, y_train, y_valid = train_test_split(
    train_feature,
    train_label,
    test_size=0.2
)

test_feature = test[feature_cols]
test_label = test[label_cols]

test_feature, test_label = make_dataset(test_feature, test_label, 20)
```

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM, Conv1D, Lambda
from keras.losses import Huber
from keras.callbacks import EarlyStopping, ModelCheckpoint
from keras.optimizers import Adam
import os

model = Sequential()
model.add(Conv1D(filters=32, kernel_size = 5, padding= "causal", activation="relu", input_shape=(train_feature.shape[1], train_feature.shape[2])))
model.add(LSTM(16, activation='tanh'))
model.add(Dense(16, activation="relu"))
model.add(Dense(1))
```

```python
loss = Huber()
optimizer = Adam(0.0005)
model.compile(loss=Huber(), optimizer=optimizer, metrics=['mse'])

history = model.fit(x_train, y_train, 
                    epochs=200, 
                    batch_size=16,
                    validation_data=(x_valid, y_valid))
```

```python
pred = model.predict(test_feature)
```

```python
plt.figure(figsize=(16, 9))
plt.plot(test_label, label='actual')
plt.plot(pred, label='prediction')
plt.legend()
plt.show()
```
